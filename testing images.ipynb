```python
import cv2
import matplotlib.pyplot as plt


def show_cv(img, title=None, cmap=None, figsize=(8, 4)):
    plt.figure(figsize=figsize)

    if img.ndim == 3:  # BGR
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cmap = None

    plt.imshow(img, cmap=cmap)
    plt.axis("off")
    if title:
        plt.title(title)
    plt.show()
```

```python
img = cv2.imread('test/Screenshots/Resident_Evil_2/image_3.png', cv2.IMREAD_COLOR)
reader = easyocr.Reader(lang_list=['en'], gpu=True)
```

```python
show_cv(img)
result = reader.readtext(img, detail=0)
print("\t".join(result))
```

```python
gray = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY)
gray = cv2.bilateralFilter(gray, d=7, sigmaColor=17, sigmaSpace=17)
show_cv(gray)
result = reader.readtext(gray, detail=0)
print("\t".join(result))
```

```python
preprocessed_image = preprocess(img.copy())
show_cv(preprocessed_image)
result = reader.readtext(preprocessed_image, detail=0)
print("\t".join(result))
```
